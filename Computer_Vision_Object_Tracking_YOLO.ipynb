```python
weight_path = 'yolov3.weights'
model = yolo_model.make_yolov3_model()
weight_reader = yolo_model.WeightReader(weight_path) # load the model weights
weight_reader.load_weights(model)  # set the model weights into the model
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

orig_image1 = cv2.imread('/content/drive/MyDrive/frame1.jpg')
orig_image2 = cv2.imread('/content/drive/MyDrive/frame2.jpg')

plt.imshow(cv2.cvtColor(orig_image1, cv2.COLOR_BGR2RGB))
plt.title('frame1')
plt.figure()
plt.imshow(cv2.cvtColor(orig_image2, cv2.COLOR_BGR2RGB))
plt.title('frame2')
```

```python
from yolo_model import one_image_detection

detected_image = yolo_model.one_image_detection(orig_image1, model, model_input_shape=(416, 416), obj_thresh=0.5, nms_thresh=0.4)
plt.imshow(cv2.cvtColor(detected_image, cv2.COLOR_BGR2RGB))
plt.figure()
```

```python
from yolo_model import one_image_detection

detected_image = yolo_model.one_image_detection(orig_image2, model, model_input_shape=(416, 416), obj_thresh=0.5, nms_thresh=0.4)
plt.imshow(cv2.cvtColor(detected_image, cv2.COLOR_BGR2RGB))
plt.figure()
```
